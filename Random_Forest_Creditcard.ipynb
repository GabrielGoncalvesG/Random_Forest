```python
# Define as variáveis independentes (X) e a variável dependente (y).
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
# Importa a função 'mutual_info_classif' do módulo 'feature_selection' do scikit-learn.
from sklearn.feature_selection import mutual_info_classif
```

```python
# Calcula a informação mútua entre cada variável em X e y
mutual_info = mutual_info_classif(X, y)

# Converte para um DataFrame para facilitar a visualização
mutual_info_series = pd.Series(mutual_info, index=X.columns)

# Ordena os valores em ordem decrescente
mutual_info_series.sort_values(ascending=False, inplace=True)

# Exibe os valores da informação mútua
print(mutual_info_series)
```

#Pré-Processamento dos dados

```python
# Importa a função 'SMOTE' do módulo 'over_sampling' do imblearn.
from imblearn.over_sampling import SMOTE
```

```python
# Aplica a técnica SMOTE para balancear o conjunto de dados.
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
print(f'Tamanho original: {X.shape}, Tamanho após SMOTE: {X_resampled.shape}')
```

```python
# Importa as funções 'train_test_split' e 'cross_val_score' do módulo 'model_selection' do scikit-learn.
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
# Divide os dados em conjuntos de treinamento e teste.
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)
```

```python
# Mostra o formato dos dados de treino e teste.
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Converte para um DataFrame
X_train = pd.DataFrame(X_train)
X_train
```

```python
# Converte para um DataFrame
X_test = pd.DataFrame(X_test)
X_test
```

#Processamento dos dados

```python
# Importa o classificador 'RandomForestClassifier' do módulo 'ensemble' do scikit-learn.
from sklearn.ensemble import RandomForestClassifier
```

```python
# Cria uma instância do classificador RandomForestClassifier com 10 árvores de decisão.
rf = RandomForestClassifier(n_estimators=10)
```

```python
# Treina o modelo com os dados de treinamento.
rf.fit(X_train, y_train)
```

```python
# Faz previsões com os dados de teste.
preds = rf.predict(X_test)
```

```python
# Importa a função 'accuracy_score' do módulo 'metrics' do scikit-learn.
from sklearn.metrics import accuracy_score
```

```python
# Calcula e imprime a acurácia do modelo.
acuracia = accuracy_score(y_test,preds)
print(f'Acurácia: {acuracia * 100}%')
```

```python
# Executa a validação cruzada para obter uma estimativa melhor da acurácia do modelo.
valid_cross = cross_val_score(rf, X_train, y_train, cv=5, scoring='accuracy')
print(f'Acurácia média é: {valid_cross.mean() * 100}%')
```

```python
# Esta função é usada para calcular a matriz de confusão.
from sklearn.metrics import confusion_matrix
```

```python
# A matriz de confusão é uma tabela que descreve o desempenho do modelo de classificação.
cm = confusion_matrix(y_test, preds)
```

```python
# Criando um mapa de calor (heatmap) da matriz de confusão usando a biblioteca seaborn.
cm_plot = sns.heatmap(cm, annot=True, cmap='Blues', fmt='d')
cm_plot.set_xlabel('Valores previstos')
cm_plot.set_ylabel('Valores reais')
cm_plot.set_title('Matriz de confusão', size=16)
plt.show()
```

#Ajuste de Hiperparâmetros

```python
# Define uma grade de parâmetros para busca aleatória.
param_dist = {
    'n_estimators': [10, 100, 200, 500],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],
    'criterion': ['gini', 'entropy']
}
```

```python
# Importa a função 'RandomizedSearchCV' do módulo 'model_selection' do scikit-learn.
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Cria uma instância do RandomizedSearchCV.
random_search = RandomizedSearchCV(RandomForestClassifier(), param_dist, n_iter=50, cv=5, scoring='accuracy', n_jobs=-1, random_state=42)
random_search.fit(X_train, y_train)
```

```python
# Imprime os melhores parâmetros encontrados.
print("Melhores parâmetros:", random_search.best_params_)
```
